## Análisis exploratorio de los datos

<center><img src='1.jpg' width=900> </center>

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/work/exams.csv')
df
```

```
#primeras 10 filas
df.head(10)
```

```
#últimas 10 filas
df.tail(10)
```

```
#descripción, resumen estadístico
df.describe()
```

```
#conteo de filas totales y duplicadas
print("Original con duplicados: ",df.id.count()," filas.")
duplicate_rows_df = df[df.duplicated()]
print("Cantidad de filas duplicadas: ",duplicate_rows_df.id.count())

#eliminación de duplicados
df=df.drop_duplicates()

#después de eliminar
print("Sin duplicados: ",df.id.count()," filas.")
```

```
#eliminar columnas irrelevantes (en este caso id)
print(df.columns)
df= df.drop(['id'], axis=1)
#verificación
print(df.columns)
```

```
#renombre de columnas
df = df.rename(columns={
    "gender":"Género",
    "race/etnicity":"Etnia",
    "parental level of education":"Nivel de educación parental",
    "lunch":"Almuerzo",
    "employed":"Tiene empleo",
    "test preparation course":"Curso de preparación para el exámen",
    "math score":"Calificación en matemática",
    "physics score":"Calificación en física",
    "chemistry score":"Calificación en química",
    "algebra_score":"Calificación en álgebra"
})
```

```
#eliminar valores nulos

#encontrar valores nulos
print(df.isnull().sum())

#eliminarlos
df=df.dropna()

#verificación
print("--------------------------------")
print(df.isnull().sum())
```

```
#detección de outliers
sns.boxplot(x=df["math score"])
plt.show()
sns.boxplot(x=df["physics score"])
plt.show()
sns.boxplot(x=df["chemistry score"])
plt.show()
sns.boxplot(x=df["algebra_score"])
plt.show()
```

```
print("Antes:", df.lunch.count(),"filas.")
Q1= df.quantile(0.25)
Q3= df.quantile(0.75)
IQR= Q3 - Q1
print(IQR)

#exclusión de outliers
df = df[~((df < (Q1-1.5*IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print("Después: ", df.lunch.count(),"filas.")
```

### Relación entre género y promedio de notas:

```
df['Average Score'] = df.mean(axis=1)
mujeres = df[df['gender']=='female']
hombres = df[df['gender']=='male']
```

```
sns.histplot(mujeres['Average Score'], color='red', alpha=.4, fill=True)
sns.histplot(hombres['Average Score'], color='blue', alpha=.4, fill=True)
plt.show()
```

```
sns.catplot(x='gender',y='Average Score',hue='test preparation course',kind='bar',data=df) 
#en hue separación por finalización del curso preparatorio para el test
```

```
print('Mujeres: ',mujeres['gender'].count())
print('Hombres: ',hombres['gender'].count())
```

La diferencia numérica entre hombres y mujeres no es muy grande, ni tampoco parecen significativas las diferencias del promedio de notas obtenidas en relación al género.

### Relación entre empleo y nota:

```
sns.boxplot(data=df,x='Average Score',y='employed',hue='gender')
```

```
tienetrabajo=df[df['employed']=='yes']
notienetrabajo=df[df['employed']=='no']
print("Cantidad de alumnos con empleo: ", tienetrabajo['employed'].count())
print("Cantidad de alumnos sin empleo: ", notienetrabajo['employed'].count())
```

La diferencia entre la cantidad de alumnos empleados y no empleados no es muy grande, ni tampoco parece incidir en el promedio de notas.

### Nota y nivel educativo de los padres:

```
sns.catplot(x='parental level of education',y='Average Score',hue='gender',aspect=2,data=df) 
```

```
sns.catplot(x='parental level of education',y='Average Score',hue='gender',kind='bar',aspect=2,data=df) 
```

```
sns.countplot(y='parental level of education',data=df)
```

```
secundarioin=df[df['parental level of education']=='some high school']
secundarioc=df[df['parental level of education']=='high school']
terciario=df[df['parental level of education']=="associate's degree"]
gradoin=df[df['parental level of education']=='some college']
grado=df[df['parental level of education']=="bachelor's degree"]
posgrado=df[df['parental level of education']=="master's degree"]

print("Cantidad con secundario incompleto: ", secundarioin['parental level of education'].count())
print("Cantidad con secundario completo: ", secundarioc['parental level of education'].count())
print("Cantidad con terciario: ", terciario['parental level of education'].count())
print("Cantidad con grado incompleto: ", gradoin['parental level of education'].count())
print("Cantidad con grado completo: ", grado['parental level of education'].count())
print("Cantidad con posgrado: ", posgrado['parental level of education'].count())
```

Pareciera haber una pequeña tendencia a mejores notas a medida que aumenta el nivel educativo de los padres. Debe tomarse en cuenta también que si bien las cantidades de alumnos cuyos padres tienen secundario incompleto, secundario completo, terciario y grado incompleto son similares, las cantidades con grado completo y posgrado son mucho menores, especialmente en el caso de este último.
